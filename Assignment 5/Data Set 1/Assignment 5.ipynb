```python
iris["SepalLenghtCm_ratio"] = iris["SepalLengthCm"] / iris["PetalLengthCm"]
```

```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(iris["SepalLengthCm"], bins=20, kde=True, color="blue")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", hue="Species", data=iris, palette="viridis")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Scatter Plot of Sepal Length vs. Petal Length")
plt.legend(title="Species")
plt.show()

```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="Species", y="SepalWidthCm", hue="Species", data=iris, palette="pastel")
plt.xlabel("Species")
plt.ylabel("Sepal Width (cm)")
plt.title("Box Plot of Sepal Width by Species")
plt.legend(title="Species")
plt.show()
```

```python
# Compute correlation matrix
corr_matrix = iris.corr(numeric_only=True)  # Ensures only numerical columns are used

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```
